```python
import matplotlib.pyplot as plt
import numpy
import pandas
```

# Confirmed COVID-19 cases
## LATAM vs EUROPE evolution

### Data from https://github.com/CSSEGISandData/COVID-19

2019 Novel Coronavirus COVID-19 (2019-nCoV) Data Repository by Johns Hopkins CSSE
This is the data repository for the 2019 Novel Coronavirus Visual Dashboard operated by the Johns Hopkins University Center for Systems Science and Engineering (JHU CSSE). Also, Supported by ESRI Living Atlas Team and the Johns Hopkins University Applied Physics Lab (JHU APL).

```python
confirmed = pandas.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
```

```python
country = confirmed.groupby('Country/Region', as_index=False).sum()
country.drop('Lat', axis=1, inplace=True)
country.drop('Long', axis=1, inplace=True)
```

```python
country.head()
```

```python
latam   = country.loc[country['Country/Region'].isin(['Argentina','Brazil','Chile','Bolivia','Colombia','Uruguay','Paraguay','Peru','Venezuela','Ecuador'])]
europe  = country.loc[country['Country/Region'].isin(['Italy','Iran','Spain'])]

compare = country.loc[country['Country/Region'].isin(['Argentina','Brazil','Spain','Chile','Italy'])]
```

# 2020-03-17 Report

## EUROPE

```python
latam.set_index("Country/Region").T.plot(figsize=(15,5))
plt.title('LATAM COVID-19 Confirmed')
plt.show()


europe.set_index("Country/Region").T.plot(figsize=(15,5))
plt.title('EUROPE COVID-19 Confirmed')
plt.show()
```

```python
nRow=0
data=[]

for index, row in compare.iterrows():
    nRow=nRow+1
    nCol=0
    nPos=0
    vec = [row['Country/Region'],numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan,numpy.nan]
    
    for (columnName, columnData) in compare.iteritems():
        nCol=nCol+1
        vec[16]=columnData.values[nRow-1]

        if nCol>1 and columnData.values[nRow-1]>0 and nPos<15:
            if (row['Country/Region'] == 'Italy' or row['Country/Region']=='Spain') and columnData.values[nRow-1] > 5: 
                nPos=nPos+1
                vec[nPos]=columnData.values[nRow-1]
            if (row['Country/Region'] != 'Italy' and row['Country/Region']!='Spain'):    
                nPos=nPos+1
                vec[nPos]=columnData.values[nRow-1]

    data.append(vec)

df = pandas.DataFrame(data, columns = ['Country','day01','day02','day03','day04','day05','day06','day07','day08','day09','day10','day11','day12','day13','day14','day15','today']) 
df.set_index('Country')


```
